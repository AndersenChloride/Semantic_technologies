### Задание 1: Сформировать новый список из всех четных элементов исходного списка
```
% Базовый случай: пустой список.
even_elements([], []).

% Если первый элемент четный, включаем его в результат.
even_elements([H|T], [H|R]) :-
    0 is H mod 2,
    even_elements(T, R).

% Если первый элемент нечетный, пропускаем его.
even_elements([H|T], R) :-
    1 is H mod 2,
    even_elements(T, R).
```

### Задание 3: Удалить все вхождения заданного элемента из списка

```
% Базовый случай: пустой список.
remove_all(_, [], []).

% Если первый элемент совпадает с заданным, пропускаем его.
remove_all(X, [X|T], R) :-
    remove_all(X, T, R).

% Если первый элемент не совпадает, включаем его в результат.
remove_all(X, [H|T], [H|R]) :-
    X \= H,
    remove_all(X, T, R).
```

### Пример запросов и интерпретации:

?- even_elements([1, 2, 3, 4, 5, 6], Result).  
Result = [2, 4, 6].  

Рекурсивно проверяется каждый элемент списка, добавляя его в результат, если он четный.

?- remove_all(3, [1, 3, 3, 4, 3, 5], Result).  
Result = [1, 4, 5].  

Проверяется каждый элемент списка, исключая совпадающие с заданным.

### Декларативная интерпретация:

**even_elements/2:**  
* Отношение между исходным списком и списком, содержащим только четные элементы из исходного. Для каждого элемента проверяется, делится ли он на 2 без остатка.  

**remove_all/3:**  
* Отношение между исходным списком, удаляемым элементом и результатирующим списком. Если элемент совпадает с заданным, он пропускается.  

### Процедурная интерпретация:

**even_elements/2:**  
* Если список пустой, результат — пустой список.  
* Если голова списка четная, она добавляется в результирующий список, затем рекурсивно обрабатывается хвост.  
* Если голова списка нечетная, она пропускается, а хвост обрабатывается рекурсивно.  

**remove_all/3:**  
* Если список пустой, результат — пустой список.  
* Если голова списка равна удаляемому элементу, она пропускается и хвост обрабатывается рекурсивно.  
* Если голова списка не равна удаляемому элементу, она включается в результирующий список и хвост обрабатывается рекурсивно.

**Задача**: Сгенерировать списки-палиндромы, состоящие из чисел из
заданного диапазона.  

### Базовый переборный алгоритм

```
% Проверка, является ли список палиндромом
is_palindrome(L) :-
    reverse(L, L).

% Генерация всех списков заданной длины
generate_list(0, _, []).
generate_list(Length, Range, [X|Rest]) :-
    Length > 0,
    member(X, Range),
    NewLength is Length - 1,
    generate_list(NewLength, Range, Rest).

% Генерация палиндромов
generate_palindromes(Min, Max, Length, Palindrome) :-
    numlist(Min, Max, Range),       % Генерация диапазона чисел
    generate_list(Length, Range, L), % Генерация списка заданной длины
    is_palindrome(L),               % Проверка, является ли список палиндромом
    Palindrome = L.
```

### Улучшение алгоритма

Для оптимизации можно сократить пространство поиска, воспользовавшись свойствами палиндрома. Например, вместо полного перебора всего списка, можно генерировать только половину элементов, а вторую половину автоматически формировать как зеркальное отражение первой.

### Оптимизированный алгоритм:

```
% Генерация половины списка
generate_half_list(0, _, []).
generate_half_list(Length, Range, [X|Rest]) :-
    Length > 0,
    member(X, Range),
    NewLength is Length - 1,
    generate_half_list(NewLength, Range, Rest).

% Генерация палиндромов с использованием половины списка
optimized_generate_palindromes(Min, Max, Length, Palindrome) :-
    numlist(Min, Max, Range),
    HalfLength is (Length + 1) // 2, % Определяем длину половины
    generate_half_list(HalfLength, Range, Half),
    mirror_list(Half, Length, Palindrome).

% Формирование полного списка палиндрома на основе его половины
mirror_list(Half, Length, Palindrome) :-
    (   Length mod 2 =:= 0
    ->  reverse(Half, Mirror),
        append(Half, Mirror, Palindrome)
    ;   reverse(Half, [_|Mirror]),
        append(Half, Mirror, Palindrome)
    ).
```

Пример использования:

?- optimized_generate_palindromes(1, 5, 3, P).  


P = [1, 2, 1] ;  
P = [1, 3, 1] ;  
P = [1, 4, 1] ;  
P = [1, 5, 1] ;  
P = [2, 1, 2] ;  
P = [2, 3, 2] ;  
P = [2, 4, 2] ;  
P = [2, 5, 2] ;  
...

Базовый алгоритм:  
Сложность: O(N^L), где N=Max−Min+1, L — длина списка, [Max, Min] - диапазон.  

Оптимизированный алгоритм:  
Сложность: O(N^(L/2)), где N=Max−Min+1, L — длина списка, [Max, Min] - диапазон.
